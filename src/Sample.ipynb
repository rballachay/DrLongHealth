```python
import json
import pandas as pd

with open('medication-SQUAD.json') as f1:
    data1 = json.load(f1)
df1 = pd.DataFrame(data1)

with open('relation-SQUAD.json') as f2:
    data2 = json.load(f2)
df2 = pd.DataFrame(data2)

print("First 10 rows of file1.json:")
print(df1.head(10))
print("\nFirst 10 rows of file2.json:")
print(df2.head(10))

# Print the size of each DataFrame
print("\nSize of DataFrame from file1.json:")
print(df1.size)
print()
print("\nSize of DataFrame from file2.json:")
print(df2.size)
```

```python
training_data = []

for _, row in df1.iterrows():
    data = row.iloc[0]  # Access the nested structure
    for paragraph in data['paragraphs']:
        for qa in paragraph['qas']:
            qa_dict = {}
            qa_dict['title'] = data['title']
            qa_dict['context'] = paragraph['context']
            qa_dict['question'] = qa['question']
            qa_dict['id'] = qa['id']
            qa_dict['answer_start'] = qa['answers'][0]['answer_start']
            qa_dict['text'] = qa['answers'][0]['text']
            training_data.append(qa_dict)

df_training = pd.DataFrame(training_data)
```

```python
from transformers import DPRContextEncoder, DPRContextEncoderTokenizer
from transformers import DPRQuestionEncoder, DPRQuestionEncoderTokenizer
import torch
from torch.optim import Adam

ctx_encoder = DPRContextEncoder.from_pretrained('facebook/dpr-ctx_encoder-single-nq-base')
ctx_tokenizer = DPRContextEncoderTokenizer.from_pretrained('facebook/dpr-ctx_encoder-single-nq-base')

ques_encoder = DPRQuestionEncoder.from_pretrained('facebook/dpr-question_encoder-single-nq-base')
ques_tokenizer = DPRQuestionEncoderTokenizer.from_pretrained('facebook/dpr-question_encoder-single-nq-base')

questions = df_training['question'].tolist()
contexts = df_training['context'].tolist()

train_question_tokens = ques_tokenizer(questions, truncation=True, padding=True, return_tensors="pt")
train_context_tokens = ctx_tokenizer(contexts, truncation=True, padding=True, return_tensors="pt")

optimizer = Adam(list(ctx_encoder.parameters()) + list(ques_encoder.parameters()), lr=1e-2)

for epoch in range(1):
    optimizer.zero_grad()

    question_embeddings = ques_encoder(**train_question_tokens).pooler_output
    context_embeddings = ctx_encoder(**train_context_tokens).pooler_output

    cosine_similarity = torch.nn.CosineSimilarity(dim=1, eps=1e-2)
    loss = 1 - cosine_similarity(question_embeddings, context_embeddings)
    loss = loss.mean()

    loss.backward()
    optimizer.step()
```

```python
from transformers import BertForQuestionAnswering, BertTokenizerFast, AdamW
from torch.utils.data import DataLoader

# Sample: use BERT base for now
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
model = BertForQuestionAnswering.from_pretrained('bert-base-uncased')

with torch.no_grad():
    context_embeddings = ctx_encoder(train_context_tokens['input_ids'], return_dict=True).pooler_output

similarity_scores = torch.matmul(question_embeddings, context_embeddings.T)
most_relevant_context_idxs = torch.argmax(similarity_scores, dim=1)
most_relevant_contexts = [contexts[i] for i in most_relevant_context_idxs]

questions = df_training['question'].tolist()
answers = df_training['text'].tolist()
inputs = tokenizer(questions, most_relevant_contexts, truncation=True, padding=True, return_tensors="pt")
start_positions = torch.tensor([context.index(answer) for context, answer in zip(most_relevant_contexts, answers)])
end_positions = torch.tensor([start + len(answer) for start, answer in zip(start_positions, answers)])
optimizer = AdamW(model.parameters(), lr=1e-2)

for epoch in range(1):
    optimizer.zero_grad()
    outputs = model(**inputs, start_positions=start_positions, end_positions=end_positions)
    loss = outputs.loss
    loss.backward()
    optimizer.step()
```
